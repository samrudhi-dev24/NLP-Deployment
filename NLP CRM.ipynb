```python
nltk.download('wordnet')
def clean_text(text):
    data = [char for char in text if char not in string.punctuation]
    data = ''.join(data)
    words = re.sub(r"(@[A-Za-z0-9]+)|([^A-Za-z0-9 \t])|(\w+:\/\/\S+)|^rt|http.+?", " ", data )
    words = words.lower()
    final_words = [wnl().lemmatize(word, pos="v") for word in words.split()]
    final_words = ' '.join(final_words)
    return final_words
```

```python
df['clean_data'] = df['text'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
bag_words = cv.fit(df['clean_data'])
```

```python
bag_words.get_feature_names()
```

```python
bag_words = bag_words.transform(df['clean_data'])
```

```python
bag_words.shape
```

```python
df.shape
```

```python
X = pd.DataFrame(data = bag_words.toarray())
```

```python
X
```

```python
df['type'].replace(to_replace = ['spam','ham'] , value = [1,0] , inplace = True)
y = df['type']
df['type'].value_counts()
```

```python
import pickle
pickle.dump(cv, open("vector_f.pkl", "wb"))
CV = pickle.load(open("vector_f.pkl", 'rb'))
CV
```

```python
var = df['clean_data'].iloc[3]
var  = [var]
```

```python
var
```

```python
bag =CV.transform(var).toarray()
bag.shape
```

```python
bag
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits=1, test_size=0.1, random_state=0)
print(sss.get_n_splits(X, y))
print(sss)

train_idx = 0
test_idx = 0
for train_idx , test_idx in sss.split(X,y):
    X_train  , X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train  , y_test = y.iloc[train_idx], y.iloc[test_idx]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
print(y.value_counts())
print(y_train.value_counts())
print(y_test.value_counts())
```

```python
from sklearn.svm import SVC
```

```python
clf = SVC(class_weight = 'balanced')


clf.fit(X_train, y_train)
```

```python
predict = clf.predict(X_train)
```

```python
predict
```

```python
expected = y_train

from sklearn import metrics
print(metrics.classification_report(expected, predict))
print(metrics.confusion_matrix(expected, predict))
```

```python
predicted = clf.predict(X_test)
expected = y_test

from sklearn import metrics
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```
